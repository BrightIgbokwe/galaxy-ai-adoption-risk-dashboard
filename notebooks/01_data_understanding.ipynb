# Defining the Prediction Target

In this project, the prediction target is customer dissatisfaction risk.

A review is considered high risk when the rating is low or the sentiment is negative. This helps us identify customers or product experiences that could reduce trust and future adoption of Galaxy AI devices.


```python
# Create a dissatisfaction risk target variable

reviews_df["dissatisfaction_risk"] = reviews_df.apply(
    lambda row: 1 if (row["rating"] <= 2 or row["sentiment"] == "Negative") else 0,
    axis=1
)

# Check the distribution of the new target variable
risk_counts = reviews_df["dissatisfaction_risk"].value_counts()

risk_counts
```

```python
# Create readable labels for the dissatisfaction risk variable

reviews_df["risk_label"] = reviews_df["dissatisfaction_risk"].map({
    1: "High Dissatisfaction Risk",
    0: "Low Dissatisfaction Risk"
})

reviews_df[["brand", "model", "rating", "sentiment", "dissatisfaction_risk", "risk_label"]].head()
```

```python
# Update Samsung reviews after creating the risk column

samsung_reviews = reviews_df[reviews_df["brand"].str.lower() == "samsung"]

# Count Samsung dissatisfaction risk
samsung_risk_counts = samsung_reviews["risk_label"].value_counts()

samsung_risk_counts
```

```python
# Calculate Samsung dissatisfaction risk percentage

samsung_risk_percentage = samsung_reviews["risk_label"].value_counts(normalize=True) * 100

samsung_risk_percentage.round(2)
```

# Understanding the Drivers of Samsung Dissatisfaction Risk

Now that we know a significant percentage of Samsung reviews show dissatisfaction risk, we need to identify which product features are most associated with that risk.

We compare feature ratings such as battery life, camera, performance, design, and display between high-risk and low-risk Samsung reviews.

```python
# Compare Samsung feature ratings by dissatisfaction risk group

feature_columns = [
    "battery_life_rating",
    "camera_rating",
    "performance_rating",
    "design_rating",
    "display_rating"
]

samsung_feature_risk_summary = samsung_reviews.groupby("risk_label")[feature_columns].mean().round(2)

samsung_feature_risk_summary
```

```python
# Compare average overall rating by risk group

samsung_rating_risk_summary = samsung_reviews.groupby("risk_label")["rating"].mean().round(2)

samsung_rating_risk_summary
```

```python
# Identify Samsung models with the highest number of high-risk reviews

samsung_high_risk = samsung_reviews[samsung_reviews["risk_label"] == "High Dissatisfaction Risk"]

high_risk_by_model = (
    samsung_high_risk["model"]
    .value_counts()
    .head(10)
)

high_risk_by_model
```

```python
# Identify countries with the highest number of high-risk Samsung reviews

high_risk_by_country = (
    samsung_high_risk["country"]
    .value_counts()
    .head(10)
)

high_risk_by_country
```

```python
# Calculate dissatisfaction risk percentage by Samsung model

model_risk_summary = (
    samsung_reviews
    .groupby("model")
    .agg(
        total_reviews=("review_id", "count"),
        high_risk_reviews=("dissatisfaction_risk", "sum"),
        average_rating=("rating", "mean"),
        average_battery_rating=("battery_life_rating", "mean"),
        average_camera_rating=("camera_rating", "mean"),
        average_performance_rating=("performance_rating", "mean"),
        average_design_rating=("design_rating", "mean"),
        average_display_rating=("display_rating", "mean")
    )
    .reset_index()
)

model_risk_summary["high_risk_percentage"] = (
    model_risk_summary["high_risk_reviews"] / model_risk_summary["total_reviews"] * 100
).round(2)

model_risk_summary = model_risk_summary.sort_values(
    by="high_risk_percentage",
    ascending=False
)

model_risk_summary.head(10)
```

```python
# Environment: VS Code Jupyter Notebook
# Create a feature gap analysis table

feature_gap_summary = samsung_feature_risk_summary.T.reset_index()

feature_gap_summary = feature_gap_summary.rename(columns={
    "index": "feature"
})

feature_gap_summary["risk_gap"] = (
    feature_gap_summary["Low Dissatisfaction Risk"] -
    feature_gap_summary["High Dissatisfaction Risk"]
).round(2)

feature_gap_summary = feature_gap_summary.sort_values(
    by="risk_gap",
    ascending=False
)

feature_gap_summary
```

```python
# Environment: VS Code Jupyter Notebook
# Calculate dissatisfaction risk percentage by country for Samsung reviews

country_risk_summary = (
    samsung_reviews
    .groupby("country")
    .agg(
        total_reviews=("review_id", "count"),
        high_risk_reviews=("dissatisfaction_risk", "sum"),
        average_rating=("rating", "mean")
    )
    .reset_index()
)

country_risk_summary["high_risk_percentage"] = (
    country_risk_summary["high_risk_reviews"] / country_risk_summary["total_reviews"] * 100
).round(2)

country_risk_summary = country_risk_summary.sort_values(
    by="high_risk_percentage",
    ascending=False
)

country_risk_summary
```

```python
# Environment: VS Code Jupyter Notebook
# Create business recommendations based on feature risk gaps

recommendation_map = {
    "battery_life_rating": "Improve battery optimization for AI features and communicate battery performance more clearly.",
    "camera_rating": "Strengthen camera experience messaging, especially for content creation and AI-enhanced photography.",
    "performance_rating": "Improve performance stability, speed, and AI processing experience across Galaxy devices.",
    "design_rating": "Protect premium design perception and improve comfort/usability for future AI devices.",
    "display_rating": "Improve display quality communication and maintain strong screen experience as a Galaxy advantage."
}

feature_gap_summary["business_recommendation"] = feature_gap_summary["feature"].map(recommendation_map)

feature_gap_summary
```

```python
# Environment: VS Code Jupyter Notebook
# Save Samsung insight tables for later Power BI and reporting

output_folder = Path("../data_cleaned")
output_folder.mkdir(exist_ok=True)

samsung_reviews.to_csv(output_folder / "samsung_reviews_with_risk.csv", index=False)
feature_gap_summary.to_csv(output_folder / "samsung_feature_gap_summary.csv", index=False)
model_risk_summary.to_csv(output_folder / "samsung_model_risk_summary.csv", index=False)
country_risk_summary.to_csv(output_folder / "samsung_country_risk_summary.csv", index=False)

print("Samsung insight tables saved successfully.")
```

# Preparing the Dataset for Prediction Modelling

The goal of the prediction model is to estimate whether a Samsung review shows high dissatisfaction risk.

To avoid data leakage, we will not use `rating` or `sentiment` as model input features because they were used to create the target variable.

```python
# Environment: VS Code Jupyter Notebook
# Prepare the Samsung dataset for prediction modelling

model_columns = [
    "age",
    "price_usd",
    "verified_purchase",
    "battery_life_rating",
    "camera_rating",
    "performance_rating",
    "design_rating",
    "display_rating",
    "review_length",
    "word_count",
    "helpful_votes",
    "country",
    "language",
    "source",
    "model",
    "dissatisfaction_risk"
]

samsung_modeling_df = samsung_reviews[model_columns].copy()

samsung_modeling_df.head()
```
